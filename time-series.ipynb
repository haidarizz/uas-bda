# Time Series 1 (Prophet)

#### Data for Training Preparation

```python
from prophet import Prophet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
pkp_df = pkp_df.rename(columns={'Tanggal':'ds', 'PERTANIAN, KEHUTANAN & PERIKANAN':'y'})
pkp_df.head()
```

Split `train_df` dan `test_df`

```python
total_rows = pkp_df.shape[0]
split_index = int(0.8 * total_rows)

# Splitting the DataFrame
df_train = pkp_df[:split_index]
df_test = pkp_df[split_index:]

# Optionally reset the index if you want
df_train.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)
```

```python
df_test
```

```python
df_train
```

```python
plt.figure(figsize=(10, 6))

# Plot df_train in blue
plt.scatter(df_train['ds'], df_train['y'], color='blue', label='Training Set', s=20)

# Plot df_test in orange
plt.scatter(df_test['ds'], df_test['y'], color='orange', label='Test Set', s=20)

plt.title('Scatter Plot of Training and Test Sets (PDB Sektor Pertanian, Kehutanan, dan Perikanan)')
plt.xlabel('Date')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
df_test
```

```python
# Create a new DataFrame with extended date range up to 2030-01-01
extended_dates = pd.date_range(start='2023-11-30', end='2030-12-2', freq='3M')  # Adjust the frequency as needed
extended_df = pd.DataFrame({'ds': extended_dates})

# Merge the new DataFrame with the original df_test and fill in 'y' values
result_df = pd.merge(extended_df, df_test, on='ds', how='left')

# Forward-fill 'y' values for dates before the start of df_test
result_df['y'] = result_df['y'].ffill()
result_df['ds'] = pd.to_datetime(result_df['ds']) + pd.DateOffset(days=1)

# Display the resulting DataFrame
result_df
```

```python
result_df = pd.concat([df_test, result_df]).reset_index(drop=True)
result_df
```

#### Prophet Model Begin

```python
model_prophet = Prophet()
```

```python
model_prophet.fit(df_train[['ds', 'y']].rename(columns={'ds': 'ds', 'y': 'y'}))
```

```python
future = pd.DataFrame(result_df['ds'])
```

```python
forecast_prophet = model_prophet.predict(future)
```

```python
forecast_prophet
```

```python
result_df['y_prophet'] = forecast_prophet['yhat']
result_df
```

```python
plt.figure(figsize=(10, 6))

# Convert 'ds' column to datetime
result_df['ds'] = pd.to_datetime(result_df['ds'])

# Plotting the actual values in blue
plt.plot(result_df[:11]['ds'], result_df[:11]['y'], color='blue', label='Actual (y)')

# Plotting the predicted values in orange
plt.plot(result_df[:11]['ds'], result_df[:11]['y_prophet'], color='orange', label='Predicted (y_prophet)')

# Adding labels and title
plt.title('Actual vs Predicted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(pkp_df['ds'], pkp_df['y'], color='blue', label='Actual (y)')
plt.plot(result_df['ds'], result_df['y_prophet'], color='orange', label='Predicted (y_prophet)')

plt.title('Actual vs Predicted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

# Time Series 2 (ARIMA)

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
import pmdarima as pm
```

```python
df_train_2 = df_train.set_index('ds')['y'].to_frame()
df_train_2
```

```python
model_sarima = pm.auto_arima(df_train_2, start_p=1, start_q=1,
                         test='adf',
                         max_p=3, max_q=3, m=12,
                         start_P=0, seasonal=True,
                         d=None, D=1, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)

model_sarima.summary()
```

```python
fitted, confint = model_sarima.predict(n_periods=200, return_conf_int=True)
index_of_fc = pd.date_range(df_train_2.index[-1], periods = 200, freq='MS')

# make series for plotting purpose
forecast_sarima = pd.Series(fitted, index=index_of_fc)
```

```python
forecast_sarima
```

```python
forecast_sarima = forecast_sarima.to_frame()
```

```python
forecast_sarima = forecast_sarima.dropna()
forecast_sarima = forecast_sarima[:40]
len(forecast_sarima)
```

```python
forecast_sarima = forecast_sarima.rename(columns={0: 'y'})
forecast_sarima
```

```python
result_df = result_df.merge(forecast_sarima, left_on='ds', right_index=True, how='left')
result_df
```

```python
result_df = result_df.rename(columns={'y_y': 'y_sarima'})
result_df = result_df.rename(columns={'y_x': 'y_real'})
result_df
```

```python
plt.figure(figsize=(10, 6))

plt.plot(pkp_df['ds'], pkp_df['y'], color='blue', label='Actual (y)')
plt.plot(result_df['ds'], result_df['y_sarima'], color='orange', label='Predicted (y_sarima)')

plt.title('Actual vs Predicted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

# Time Series 3 (Exponential Smoothing)

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
# create class
model_ses = SimpleExpSmoothing(df_train_2['y'])
model_ses_fit = model_ses.fit(optimized = True)

print('Smoothing Level', np.round(model_ses_fit.params['smoothing_level'], 4))
print('Initial Level',   np.round(model_ses_fit.params['initial_level'], 4))
```

```python
forecast_ses = model_ses_fit.forecast(len(result_df['y_real']))
```

```python
forecast_ses = forecast_ses.to_frame()
forecast_ses
```

```python
forecast_ses = forecast_ses.rename(columns={0: 'y'})
forecast_ses
```

```python
result_df = result_df.merge(forecast_ses, left_on='ds', right_index=True, how='left')
result_df
```

```python
result_df = result_df.rename(columns={'y': 'y_ses'})
result_df
```

```python
plt.figure(figsize=(10, 6))

plt.plot(pkp_df['ds'], pkp_df['y'], color='blue', label='Actual (y)')
plt.plot(result_df['ds'], result_df['y_ses'], color='orange', label='Predicted (y_ses)')

plt.title('Actual vs Predicted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
